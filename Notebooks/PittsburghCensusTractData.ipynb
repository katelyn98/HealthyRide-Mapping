```python
import pandas as pd
```

```python
cen_combo_df = pd.read_excel("../Data/CensusTractData/ACDDP5_combined_cleaned.xlsx")
```

```python
cen_combo_df
```

```python
cen_combo_df.columns
```

```python
for col in cen_combo_df:
    if 'Unnamed:' in col and col is not 'Unnamed: 0': 
        cen_combo_df = cen_combo_df.drop(columns=[col])
    elif 'Census' in col:
        cen_num = col.split(' ')[2]
        cen_num = cen_num[:-1]
        if len(cen_num) == 3:
            cen_num = '0' + cen_num + '00'
        elif len(cen_num) == 4:
            cen_num = cen_num + '00'
        cen_combo_df = cen_combo_df.rename(columns={col: cen_num})
    
cen_combo_df
```

### ACS 5-Year Estimates Data Profile 2019
Final dataframe is cen_combo_df

```python
cen_combo_df = cen_combo_df.drop([0])
cen_combo_df
```

### Commuting to Work
Limit ACS 2019 5-year estimate to the commuting to work category for all chosen census tracts in allegheny county

```python
transit_cen_df = cen_combo_df[19:27]
transit_cen_df = transit_cen_df.reset_index()
transit_cen_df = transit_cen_df.drop(columns=['index'])
transit_cen_df = transit_cen_df.T
transit_cen_df = transit_cen_df.reset_index()
transit_cen_df = transit_cen_df.rename(columns={'index':'tractce10', 0: 'Workers_16yrs+', 1: 'own_alone', 
                                                2: 'carpool', 3: 'public_transit', 4: 'walked', 5: 'other',
                                               6: 'wfh', 7: 'mean_travel_time_to_work'})
transit_cen_df
```

```python
transit_cen_df.columns
```

```python
## save dataframe to csv to use in larger analysis
transit_cen_df.to_csv('../Data/CensusTractData/commuting_data_pa.csv')
```

### Employment Status and Population
Limit ACS 2019 5-year estimate to the employment status categoru for all chosen census tracts in allegheny county

```python
emppop_cen_df = cen_combo_df[1:18]
emppop_cen_df = emppop_cen_df.reset_index()
emppop_cen_df = emppop_cen_df.drop(columns=['index'])
emppop_cen_df = emppop_cen_df.T
emppop_cen_df = emppop_cen_df.reset_index()
emppop_cen_df = emppop_cen_df.rename(columns={'index':'tractce10', 0:'population_16yrs+', 1: 'in_labor_force',
                                             2: 'civ_labor_force', 3: 'employed', 4: 'unemployed', 
                                             5: 'armed forces', 6: 'not_in_labor_force', 7: 'civ_labor_force',
                                             8: 'unemployement_rate', 9: 'females_16yr+', 10: 'in_labor_force',
                                             11: 'civ_labor_force', 12: 'employed'})
emppop_cen_df
```

```python
## save dataframe to csv to use in larger analysis
emppop_cen_df.to_csv('../Data/CensusTractData/employment_data_pa.csv')
```

### Income and Benefits
Limit ACS 2019 5-year estimate to the employment status categoru for all chosen census tracts in allegheny county

```python
incben_cen_df = cen_combo_df[28:72]
incben_cen_df = incben_cen_df.reset_index()
incben_cen_df = incben_cen_df.drop(columns=['index'])
incben_cen_df = incben_cen_df.T
incben_cen_df = incben_cen_df.reset_index()
incben_cen_df = incben_cen_df.rename(columns={'index':'tractce10', 1: 'total_households', 2: '<$10K', 
                                              3: '$10K-15K', 4: '$15K-25K', 5:'$25K-35K', 6: '$35K-50K',
                                             7: '$50K-75K', 8: '$75K-100K', 9:'$100K-150K', 
                                             10: '$150K-200K', 11: '>$200K', 12: 'median_hh_income',
                                             13: 'mean_hh_income', 43: 'med_earnings_female_full-time',
                                             42: 'med_earnings_male_full-time'})
incben_cen_df
```

```python
## save dataframe to csv to use in larger analysis
incben_cen_df.to_csv('../Data/CensusTractData/income_data_pa.csv')
```
